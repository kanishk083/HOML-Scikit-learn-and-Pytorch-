# auto-differentiation feature.

AutoGrad

```python
x = torch.tensor(5.0, requires_grad=True)
f = x ** 2
f
```

```python
f.backward()
x.grad
```

```python
learning_rate = 0.1
with torch.no_grad():
  x -= learning_rate * x.grad #Gradient descent
```

```python
x
```

```python
import torch
learning_rate=0.1
x =torch.tensor(6.7)
y= torch.tensor(0.0)
```

```python
w = torch.tensor(1.0,requires_grad=True)
b = torch.tensor(0.0,requires_grad=True)
```

```python
w
```

```python
b
```

```python
z=w*x + b
z
```

```python
y_pred = torch.sigmoid(z)
y_pred
```

```python
import torch.nn as nn

criterion = nn.BCELoss()
loss = criterion(y_pred, y)
loss
```

```python
loss.backward()
```

```python
print(w.grad)
print(b.grad)
```
